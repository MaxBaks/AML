```python
adult = adult.replace(['?'], np.nan)
adult.info()
```

```python
adult = adult.dropna()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

features = adult[['age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country' ]]
X = pd.get_dummies( features, columns = [ 'workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country' ] )
X.head()

```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit( adult['income'] )
```

```python
y = pd.Series( data = le.transform( adult['income'] ) )
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python
#  
model = LogisticRegression()

model.fit( x_train, y_train )
```

```python
model.score(x_train, y_train)
```

```python
prediction = model.predict(x_test)
model.score(x_test, y_test)
```

```python
#SVM
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel = "rbf")
```

```python
classifier.fit(x_train, y_train)
```

```python
score_1 = classifier.score(x_train, y_train)
score_1
```

```python
score_1 = classifier.score(x_test, y_test)
score_1
```
